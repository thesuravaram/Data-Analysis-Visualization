# **Descriptive** **Statistics**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats
```

```python
data = {
    "Class": ["A"]*10 + ["B"]*10,
    "Scores": [78, 85, 90, 88, 76, 95, 89, 84, 91, 87, 72, 70, 68, 75, 80, 74, 77, 69, 73, 71]
}
df = pd.DataFrame(data)

print(df.groupby("Class")["Scores"].describe())
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Class", y ="Scores", data=df, hue="Class", palette="Set2")
plt.title("Boxplot: Scores by Class")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df, x="Scores", hue="Class", bins=8, kde=True, palette="Set1")
plt.title("Histogram: Score Distribution by Class")
plt.show()
```

```python
cl = 0.95

ci = stats.t.interval(cl, df=len(df["Scores"])-1, loc=np.mean(df["Scores"]), scale=stats.sem(df["Scores"]))

print(f"Confidence Interval: {ci}")
```
